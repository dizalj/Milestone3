```python
import os
import yaml
import subprocess

def initialize_feast_feature_store():
    """Initialize Feast feature store in Google Colab."""
    feature_repo_path = "/content/feature_repo"
    feature_store_yaml = os.path.join(feature_repo_path, "feature_store.yaml")

    if not os.path.exists(feature_repo_path):
        print(f"Feature store repo does not exist. Creating at {feature_repo_path}...")
        os.makedirs(feature_repo_path)

        redis_status = subprocess.run(["redis-server", "--version"], capture_output=True, text=True)
        if redis_status.returncode != 0 or "Redis server" not in redis_status.stdout:
            print("Installing Redis...")
            subprocess.run(["apt-get", "update", "-y"])
            subprocess.run(["apt-get", "install", "redis-server", "-y"])
            print("Starting Redis server...")
            subprocess.run(["redis-server", "--daemonize", "yes"])

        feature_store_config = {
            "project": "recipe_feature_store",
            "online_store": {
                "type": "redis",
                "host": "localhost",
                "port": 6379,
            },
            "offline_store": {
                "type": "parquet",
                "path": "/content/offline_data",
            }
        }

        with open(feature_store_yaml, "w") as f:
            yaml.dump(feature_store_config, f)
        print("Feature store configuration created.")
    else:
        print(f"Feature store repo already exists at {feature_repo_path}.")

initialize_feast_feature_store()
```

```python
from pymongo import UpdateOne

@step
def data_ingestion() -> pd.DataFrame:
    """Ingests raw data from CSV, stores only new unique entries in MongoDB based on the name of the ingredient, and retrieves all stored data."""
    print("Ingesting...")
    df = pd.read_csv("data.csv")

    if not df.empty:
        records = df.to_dict(orient="records")

        bulk_operations = []

        for record in records:
            bulk_operations.append(
                UpdateOne(
                    {"Ingredient": record["Ingredient"]},
                    {"$setOnInsert": record},
                    upsert=True
                )
            )

        # Perform the bulk upsert operation
        if bulk_operations:
            collection.bulk_write(bulk_operations)

    # Retrieve all data from MongoDB
    data = pd.DataFrame(list(collection.find({}, {"_id": 0})))
    return data
```

```python
@step
def data_preprocessing(df: pd.DataFrame) -> pd.DataFrame:
    print("Preprocessing...")

    if "Ingredient" in df.columns:
        df["Ingredient"] = df["Ingredient"].str.lower()

    if "Description" in df.columns:
        df["Description"] = df["Description"].str.lower()

    if "Category" in df.columns:
        df["Category"] = df["Category"].str.lower()

    if "Substitutes" in df.columns:
        df["Substitutes"] = df["Substitutes"].apply(lambda x: parse_substitutes(x))

    df = df.dropna()

    return df

def parse_substitutes(substitute_list):
    parsed_data = []

    if isinstance(substitute_list, str):
        substitute_list = ast.literal_eval(substitute_list)

    for substitute in substitute_list:
        substitute_data = {
            "Substitute": substitute.get("Substitute", ""),
            "Equivalent": substitute.get("Equivalent", ""),
            "Note": substitute.get("Note", ""),
        }
        parsed_data.append(substitute_data)

    return parsed_data
```

```python
@step
def data_validation(df: pd.DataFrame) -> pd.DataFrame:
    print("Validating...")
    required_columns = ["Ingredient", "Description", "Category", "Substitutes"]
    for column in required_columns:
        if column not in df.columns:
            print(f"Error: Missing required column: {column}")
            return df

    inferred_schema = {col: str(df[col].dtype) for col in df.columns}
    print("Inferred Schema:")
    for col, dtype in inferred_schema.items():
        print(f"{col}: {dtype}")

    missing_values = df.isnull().sum()
    for col, count in missing_values.items():
        if count > 0:
            print(f"Warning: {count} missing values in '{col}' column.")

    invalid_substitute_format = 0
    for index, substitute_list in df["Substitutes"].items():
        if isinstance(substitute_list, str):
            try:
                substitute_list = ast.literal_eval(substitute_list)
            except Exception as e:
                print(f"Error: Invalid string format in 'Substitutes' at index {index}: {e}")
                invalid_substitute_format += 1
                continue

        if not isinstance(substitute_list, list) or not all(isinstance(sub, dict) for sub in substitute_list):
            invalid_substitute_format += 1
            print(f"Error: Invalid substitute format at index {index}.")

    if invalid_substitute_format > 0:
        print(f"Warning: {invalid_substitute_format} invalid substitute formats detected.")

    # Data Statistics
    print("Data Statistics:")
    print(df.describe())

    print("Data validation complete.")
    return df
```

```python
@step
def data_versioning(df: pd.DataFrame):
    """Save the processed dataset and version it using DVC."""
    print("Starting data versioning with DVC...")

    timestamp = time.strftime("%Y%m%d-%H%M%S")
    versioned_filename = f"processed_data_{timestamp}.csv"

    df.to_csv(versioned_filename, index=False)
    time.sleep(2)

    if not os.path.exists(versioned_filename):
        print(f"Error: File {versioned_filename} not found after saving.")
        return

    try:
        print("Performing DVC versioning...")
        subprocess.run(["dvc", "add", versioned_filename], check=True)
        subprocess.run(["git", "add", f"{versioned_filename}.dvc"], check=True)
        subprocess.run(["git", "commit", "-m", f"Versioned dataset: {timestamp}"], check=True)
        subprocess.run(["dvc", "push"], check=True)

        print(f"Data saved as {versioned_filename} and successfully versioned with DVC.")

    except subprocess.CalledProcessError as e:
        print(f"DVC versioning failed: {e}")
```

```python
from feast import FeatureStore, Entity, Feature, FeatureView, ValueType
import pandas as pd
import numpy as np
import os
import time

@step
def feature_store_integration(df: pd.DataFrame):
    print("Feature Store Integration...")
    df["Ingredient_Embedding"] = df["Ingredient"].apply(lambda x: nlp.encode(x) if isinstance(x, str) else np.zeros(384))
    df["Description_Embedding"] = df["Description"].apply(lambda x: nlp.encode(x) if isinstance(x, str) else np.zeros(384))

    # Define Feast Entities and Features
    ingredient_entity = Entity(name="ingredient", value_type=ValueType.STRING)

    category_feature = Feature(name="category", dtype=ValueType.STRING)
    ingredient_embedding_feature = Feature(name="ingredient_embedding", dtype=ValueType.FLOAT_VECTOR, shape=(384,))
    description_embedding_feature = Feature(name="description_embedding", dtype=ValueType.FLOAT_VECTOR, shape=(384,))

    # Define Feature View
    recipe_feature_view = FeatureView(
        name="recipe_features",
        entities=[ingredient_entity],
        features=[category_feature, ingredient_embedding_feature, description_embedding_feature],
        batch_source=df
    )

    # Define Feature Repository Path
    feature_repo_path = "feature_repo"
    feature_store_yaml = os.path.join(feature_repo_path, "feature_store.yaml")

    if os.path.exists(feature_repo_path) and os.path.isfile(feature_store_yaml):
        try:
            store = FeatureStore(repo_path=feature_repo_path)
            store.apply([recipe_feature_view])
            store.materialize()

            print("Features stored successfully in Feast.")
        except Exception as e:
            print(f"Error occurred while accessing Feast feature store: {e}")
    else:
        print(f"Feature store repository or configuration file missing at {feature_repo_path}")
```

```python
# Define the ZenML pipeline
@pipeline
def data_pipeline():
    df = data_ingestion()
    df = data_preprocessing(df)
    df = data_validation(df)
    feature_store_integration(df)
    data_versioning(df)

data_pipeline()
```
